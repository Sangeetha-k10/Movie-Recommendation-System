data analysing II

```python
movie.iloc[0]['genres']
```

```python
import ast
def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
movie.head(2)
```

```python
movie.iloc[0]['keywords']
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
movie.head(2)
```

```python
import ast
def convert_cast(text):
    l=[]
    c=0
    for i in ast.literal_eval(text):
        if c<3:
            l.append(i['name'])
        c+=1
    return l
```

```python
movie['cast']=movie['cast'].apply(convert_cast)
```

```python
movie.head(2)
```

```python
import ast
def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movie['crew']=movie['crew'].apply(fetch_director)
```

```python
movie.head(2)
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split())
```

```python
movie.head(2)
```

```python
def remove_space(word):
    l=[]
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movie['cast']=movie['cast'].apply(remove_space)
movie['crew']=movie['crew'].apply(remove_space)
movie['genres']=movie['genres'].apply(remove_space)
movie['keywords']=movie['keywords'].apply(remove_space)
```

```python
movie.head(2)
```

```python
movie['release_year'] = pd.to_datetime(movie['release_date'], errors='coerce').dt.year
movie.drop(columns=['release_date'], inplace=True)
```

```python
movie.head(2)
```

```python
movie['tags']=movie['overview']+movie['genres']+movie['keywords']+movie['cast'] + movie['crew']
```

```python
movie.head(2)
```

```python
new_df=movie[['movie_id','title','release_year','tags','vote_average']]
```

```python
new_df.head(2)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(2)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stems(text):
    if isinstance(text, str):   
        return " ".join([ps.stem(word) for word in text.split()])
    return ""
```

```python
new_df['tags']=new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

index of the movie

```python
new_df[new_df['title']=='Spider-Man'].index[0]
```

```python
def recommend(movie):
    index=new_df[new_df['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:  x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```
